```python
print("\n=== 5. ВЕКТОРНЫЕ ОПЕРАЦИИ И СТАТИСТИКА ===")

# Создаем два массива для операций
array_a = np.array([1, 2, 3, 4, 5])
array_b = np.array([10, 20, 30, 40, 50])

print("Массив A:", array_a)
print("Массив B:", array_b)

# Поэлементное сложение массивов
addition = array_a + array_b
print("Поэлементное сложение A + B:", addition)

# Поэлементное умножение массивов
multiplication = array_a * array_b
print("Поэлементное умножение A * B:", multiplication)

# Математические операции с массивами
squared = array_a ** 2
print("Квадрат элементов массива A:", squared)

# Базовые статистические операции
print("\n--- СТАТИСТИЧЕСКИЕ ХАРАКТЕРИСТИКИ ---")
print("Массив для статистики:", array_a)

mean_value = np.mean(array_a)  # Среднее значение
print("Среднее значение:", mean_value)

sum_value = np.sum(array_a)    # Сумма всех элементов
print("Сумма всех элементов:", sum_value)

max_value = np.max(array_a)    # Максимальное значение
print("Максимальный элемент:", max_value)

min_value = np.min(array_a)    # Минимальное значение
print("Минимальный элемент:", min_value)

std_value = np.std(array_a)    # Стандартное отклонение
print("Стандартное отклонение:", std_value)

# Статистика для двумерного массива pypy
matrix_stats = np.random.rand(4, 3)
print("\nДвумерный массив для статистики:")
print(matrix_stats)
print("Среднее по всему массиву:", np.mean(matrix_stats))
print("Среднее по столбцам:", np.mean(matrix_stats, axis=0))
print("Среднее по строкам:", np.mean(matrix_stats, axis=1))
```
